```python
import seaborn as sb
%matplotlib inline
```

```python
dataset = sb.load_dataset('tips')
dataset.head()
```

```python
sb.lmplot(x='total_bill', y='tip', data=dataset, hue='sex', markers=['v', 'o'],
          scatter_kws={'s': 1})
```

```python
sb.lmplot(x='total_bill', y='tip', data=dataset, col='sex', row='time',
          scatter_kws={'s': 1})
```

```python
sb.lmplot(x='total_bill', y='tip', data=dataset, hue='sex', col='day',
          scatter_kws={'s': 5}, aspect=0.5, height=10)
```
